**Filter on Multiple Conditions**

- and : &
- or : |

```python
(df['PhoneService'] == 'Yes') & (df['PaperlessBilling'] == 'Yes')
```

```python
df[ (df['PhoneService'] == 'Yes') & (df['PaperlessBilling'] == 'Yes')]
```

**Check if we have more number of Churn customers for month to month contract or it is for One Year Contract?**

```python
df[(df['Contract'] == 'Month-to-month') & (df['Churn']=='Yes')].shape
```

```python
df[(df['Contract'] == 'One year') & (df['Churn']=='Yes')].shape
```

**Group/Summarising**

```python
df.head()
```

```python
df.groupby('Churn')
```

```python
df.groupby('Churn').agg( {'customerID':'count', 'tenure':'mean'} )
```

**Find out which kind of payment method is having more churn customers?**

```python
df[df['Churn']=='Yes'].groupby('PaymentMethod').agg({'customerID':'count'})
```

```python
df.groupby( ['Contract', 'Churn'] ).agg({'customerID':'count'})
```

```python
pd.pivot_table(df, index = 'Churn', columns = 'Contract', values = 'customerID' , aggfunc = 'count')
```

**check out of the churn or non-churn customers which kind of customer using  a particular contract types is having a higher avg tenure?**

```python
pd.pivot_table(df, index = 'Churn', columns = 'Contract', values = 'tenure' , aggfunc = 'mean')
```
